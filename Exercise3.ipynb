```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
data = pd.read_csv(url, sep = '\t')
data.head()
```

```python
transform_price = lambda x:float(x[1:-1])
data.item_price = data.item_price.apply(transform_price) 
```

```python
filtered_data = data.drop_duplicates(['item_name','quantity','choice_description'])
filtered_data
```

```python
one_product = filtered_data[filtered_data.quantity == 1]
one_product
```
