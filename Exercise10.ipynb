```python
import numpy as np
import pandas as pd
```

```python
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")

print(cars1.head())
print(cars2.head())
```

```python
cars1.info()
```

```python
cars1 = cars1.loc[:, 'mpg':'car']
cars1
```

```python
print(cars1.shape)
print(cars2.shape)
```

```python
cars = cars1.append(cars2)
cars
```

```python
nr_owners = np.random.randint(low=15000, high=73001, size=398)
nr_owners
```

```python
nr_owners = pd.Series(nr_owners)
nr_owners
```

```python
cars["owners"] = nr_owners
```

```python
cars
```

```python
nr_owners = pd.DataFrame(nr_owners, columns=["owner"])
nr_owners
```

```python
concat = pd.concat([cars1, nr_owners], axis=1)
concat
```
